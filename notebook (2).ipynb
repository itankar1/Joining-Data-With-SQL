## JOIN THE CITIES AND COUNTRIES DATAS

SELECT cities.name as city, countries.country_name as country, region 

FROM cities

INNER JOIN countries

ON cities.country_code = countries.code;

```python
SELECT cities.name as city, countries.country_name as country, region 
FROM cities
INNER JOIN countries
ON cities.country_code = countries.code;
```

## JOIN THE COUNTRIES AND ECONOMIES DATAS

SELECT c.code AS country_code, e.year, e.inflation_rate, c.country_name

FROM countries AS c

INNER JOIN economies AS e

ON c.code = e.code;

```python
SELECT c.code AS country_code, e.year, e.inflation_rate, c.country_name
FROM countries AS c
INNER JOIN economies AS e
ON c.code = e.code;
```

## JOIN THE CITIES AND COUNTRIES DATAS

SELECT 
    c1.name AS city,
    
    c1.country_code,
    
    c2.country_name AS country,
    
    c2.region,
    
    c1.city_proper_pop
    
FROM cities AS c1

INNER JOIN countries AS c2

ON c1.country_code = c2.code

ORDER BY c1.country_code DESC;

```python
SELECT 
    c1.name AS city,
    c1.country_code,
    c2.country_name AS country,
    c2.region,
    c1.city_proper_pop
FROM cities AS c1
INNER JOIN countries AS c2
ON c1.country_code = c2.code
ORDER BY c1.country_code DESC;
```
